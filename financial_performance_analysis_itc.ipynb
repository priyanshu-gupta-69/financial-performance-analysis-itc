# Financial Performance & Ratio Analysis of ITC Ltd

## Objective
The objective of this project is to analyze the financial performance of ITC Ltd over a five-year period (FY 2020–FY 2024) using financial data and ratio analysis.

The analysis focuses on understanding:
- Revenue and profit trends
- Profitability of the company
- Liquidity position
- Efficiency in using assets

Python is used as a tool to organize the data, calculate ratios, and visualize trends. The interpretation is based on basic finance and accounting concepts.


```python
import pandas as pd

df = pd.read_csv('/content/financials.csv')
df
```

```python
df.info()
df.isnull().sum()
```

## Data Preparation

The financial data used in this project has been collected from publicly available financial statements of ITC Ltd.  
Before analysis, the data was cleaned to ensure numerical consistency and accuracy.

All monetary values are expressed in ₹ Crore.  
This step ensures that the dataset is reliable and suitable for further financial analysis.


```python
def net_profit_margin(net_profit, revenue):
    return (net_profit / revenue) * 100

def return_on_equity(net_profit, equity):
    return (net_profit / equity) * 100

def current_ratio(current_assets, current_liabilities):
    return current_assets / current_liabilities

def asset_turnover(revenue, total_assets):
    return revenue / total_assets
```

```python
df['net_profit_margin'] = net_profit_margin(df['net_profit'], df['revenue'])
df['roe'] = return_on_equity(df['net_profit'], df['equity'])
df['current_ratio'] = current_ratio(df['current_assets'], df['current_liabilities'])
df['asset_turnover'] = asset_turnover(df['revenue'], df['total_assets'])

df
```

```python
df[['net_profit_margin', 'roe', 'current_ratio', 'asset_turnover']] = (
    df[['net_profit_margin', 'roe', 'current_ratio', 'asset_turnover']].round(2)
)

df
```

## Financial Ratio Calculation

Financial ratios are calculated to convert raw financial numbers into meaningful indicators of business performance.

The following categories of ratios are used:
- **Profitability ratios** to measure how efficiently the company generates profit
- **Liquidity ratios** to assess short-term financial stability
- **Efficiency ratios** to evaluate how effectively assets are utilized

These ratios help in comparing performance across different years in a standardized manner.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(df['year'], df['revenue'], marker='o')
plt.xlabel('Year')
plt.ylabel('Revenue (₹ Crore)')
plt.title('ITC Revenue Trend (2020–2024)')
plt.show()
```

### Revenue Trend Analysis

The revenue trend shows how the company’s total income has changed over the five-year period.  
ITC’s revenue declined during FY 2021 due to the impact of COVID-19, followed by a steady recovery in subsequent years.

This upward trend after FY 2021 indicates business resilience and stable demand across ITC’s core segments.


```python
plt.figure()
plt.plot(df['year'], df['net_profit'], marker='o')
plt.xlabel('Year')
plt.ylabel('Net Profit (₹ Crore)')
plt.title('ITC Net Profit Trend (2020–2024)')
plt.show()
```

### Net Profit Trend Analysis

Net profit represents the company’s earnings after all expenses and taxes.  
The net profit trend shows a strong increase over time, particularly after FY 2021.

The fact that net profit grows faster than revenue suggests improved cost control and better operational efficiency.


```python
plt.figure()
plt.plot(df['year'], df['net_profit_margin'], marker='o')
plt.xlabel('Year')
plt.ylabel('Net Profit Margin (%)')
plt.title('ITC Net Profit Margin Trend')
plt.show()
```

### Net Profit Margin Analysis

Net profit margin indicates how much profit the company earns for every ₹100 of revenue.  
ITC maintains consistently high net profit margins throughout the period analyzed.

This reflects strong pricing power, efficient operations, and effective cost management.


```python
plt.figure()
plt.plot(df['year'], df['roe'], marker='o')
plt.xlabel('Year')
plt.ylabel('ROE (%)')
plt.title('ITC Return on Equity (ROE) Trend')
plt.show()
```

### Return on Equity (ROE) Analysis

Return on Equity (ROE) measures how efficiently the company uses shareholders’ funds to generate profit.  
The upward trend in ROE indicates improving returns for shareholders over time.

This suggests that ITC is using its equity capital efficiently.


```python
plt.figure()
plt.plot(df['year'], df['current_ratio'], marker='o')
plt.xlabel('Year')
plt.ylabel('Current Ratio')
plt.title('ITC Current Ratio Trend')
plt.show()
```

### Liquidity Analysis (Current Ratio)

The current ratio measures the company’s ability to meet short-term obligations using short-term assets.  
ITC’s current ratio remains well above 1 across all years, indicating a strong liquidity position.

While high liquidity reflects financial safety, extremely high values may also indicate conservative use of resources.


```python
plt.figure()
plt.plot(df['year'], df['total_assets'], marker='o')
plt.xlabel('Year')
plt.ylabel('Total Assets (₹ Cr)')
plt.title('Total Assets Trend')
plt.show()
```
